```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The dataset I decided to use was called "Student Performance" from the website Kaggle. I chose this becasue Kaggle has really quality datasets for those who are beginner to intermediate oders so very easy to use.

These are the columns and what they mean:

- gender: column contains values "male" or "female"
- race/ethnicity: column contains values "group A", "group B", "group C", or "group D"
- parental level of education: column contains values "some college", "associate's degree", "high school", "some high school", or "bachelor's degree"
- lunch: column contains values "standard" or "free/reduced"
- test preparation course: column contain values "none" or "completed"
- math score: column contains numerical values 0 to 100
- reading score: column contains numerical values 0 to 100
- writing score: column contains numerical values 0 to 100

```python
df = pd.read_csv("StudentsPerformanceData.csv")
df
```

```python
# Graph 1: Curious about the Distribution of Math Scores
plt.figure(figsize=(8, 6))
sns.histplot(df['math score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Graph 2: Creatung a Boxplot of Reading Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='reading score', data=df, palette='pastel')
plt.title('Boxplot of Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()
```

```python
# Graph 3: Finding The Barplot of Average Writing Scores by Parental Level of Education
plt.figure(figsize=(10, 6))
sns.barplot(x='parental level of education', y='writing score', data=df, palette='muted')
plt.title('Average Writing Scores by Parental Level of Education')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Writing Score')
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.show()
```

```python
# Graph 4: Creating a Scatterplot of Math vs Reading Scores with Lunch as Hue
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', hue='lunch', data=df, palette='dark')
plt.title('Scatterplot of Math vs Reading Scores with Lunch')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.legend(title='Lunch')
plt.show()
```
